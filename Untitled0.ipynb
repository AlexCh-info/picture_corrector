```python
from fastai.vision.all import *
from fastai.vision.augment import aug_transforms
import torch.nn as nn
from pathlib import Path
```

```python
path = Path('/content/drive/MyDrive/ml/vk/data')
```

```python
def get_target_path(fn):
    """
    Превращает input имя в target имя, убирая суффикс перед расширением.
    Пример: ..._MG_1828_N1.jpg -> ..._MG_1828.jpg
    """
    fname = fn.name
    stem, ext = fname.rsplit('.', 1)  # Разделяем имя и расширение (например, 'name_N1' и 'jpg')

    # Находим позицию последнего подчеркивания
    last_underscore = stem.rfind('_')

    if last_underscore != -1:
        # Отрезаем всё после последнего подчеркивания
        clean_stem = stem[:last_underscore]
    else:
        clean_stem = stem

    # Собираем имя обратно и меняем папку на target
    new_fname = f"{clean_stem}.{ext.lower()}"
    new_path = fn.parent.parent / 'GT_IMAGES' / new_fname

    return new_path
```

```python
files = get_image_files(path/'INPUT_IMAGES')
```

```python
blocks = (ImageBlock, ImageBlock)
```

```python
photo_redactor = DataBlock(
    blocks=blocks,
    get_items = lambda _: files,
    get_y = get_target_path,
    splitter = RandomSplitter(valid_pct=0.2, seed=42),
    item_tfms = Resize(256),
    batch_tfms=[*aug_transforms(flip_vert=True, max_rotate=10),
                Normalize.from_stats(*imagenet_stats)]
)
```

```python
dls = photo_redactor.dataloaders(path, bs=8)
```

```python
dls.show_batch(max_n=4)
```

```python
learn = unet_learner(dls, resnet18, n_out=3, loss_func=L1LossFlat()).to_fp16()
```

```python
batch = dls.one_batch()
print(f'Input: {batch[0].shape}')
print(f'Target: {batch[1].shape}')
```

```python
learn.fine_tune(2)
```

```python
learn.export('/content/drive/MyDrive/ml/photo_enhancer.pkl')
```

```python
class CombinedLoss(nn.Module):
    def __init__(self, l1_weight=1.0, perceptual_weight=0.1):
        super().__init__()
        from fastai.losses import L1LossFlat
        self.l1 = L1LossFlat()
        self.l1_weight = l1_weight
        self.perceptual_weight = perceptual_weight

        # Perceptual loss
        from torchvision.models import vgg16
        vgg = vgg16(pretrained=True).features[:16]
        vgg.eval()
        for param in vgg.parameters():
            param.requires_grad = False
        self.vgg = vgg

    def forward(self, pred, target):
        l1_loss = self.l1(pred, target)

        # Perceptual loss
        pred_features = self.vgg(pred)
        target_features = self.vgg(target)
        perceptual_loss = nn.functional.l1_loss(pred_features, target_features)

        return self.l1_weight * l1_loss + self.perceptual_weight * perceptual_loss
```
